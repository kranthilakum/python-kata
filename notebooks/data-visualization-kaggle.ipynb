## Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
titanic_data = pd.read_csv("../data/titanic.csv")
titanic_data.head()
```

```python
titanic_data.describe()
```

```python
# Number of minors on board
(titanic_data['Age'] < 18).sum()
```

```python
sns.scatterplot(x=titanic_data["Age"], y=titanic_data["Fare"])
plt.title("Fare Paid and Age of Passengers")
plt.xlabel("Age")
```

```python
# line chart showing how fares change with age
sns.lineplot(x=titanic_data["Age"], y=titanic_data["Fare"])
plt.title("Fare Paid and Age of Passengers")
plt.xlabel("Age")
```

```python
# Line chart on minors (age < 18) showing how fares change with age
sns.lineplot(x=titanic_data["Age"][titanic_data["Age"] < 18], y=titanic_data["Fare"][titanic_data["Age"] < 18])
plt.title("Fare Paid and Age of Minors")
plt.xlabel("Age")
```

```python
fifa_filepath = "../data/fifa_player_data.csv"

# Read the file into a variable fifa_data
fifa_player_data = pd.read_csv(fifa_filepath, index_col="Nationality", parse_dates=True)
fifa_player_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing player ages for each country
sns.lineplot(data=fifa_player_data, x=fifa_player_data.index, y="Age")
# legend on x axis
plt.xticks(rotation=90)
```

```python
# bar chart comparing the number of survivors and non-survivors by passenger class
sns.barplot(x=titanic_data["Pclass"], y=titanic_data["Survived"])
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
```

```python
# histogram of ages to explore the age distribution of the passengers
sns.histplot(titanic_data["Age"], kde=False)
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
```

```python
# box plot comparing the age distributions of passengers in first, second, and third class
sns.boxplot(x=titanic_data["Pclass"], y=titanic_data["Age"])
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
```

```python
# box plot comparing the age distributions of survivors and non-survivors
sns.boxplot(x=titanic_data["Survived"], y=titanic_data["Age"])
plt.title("Age Distribution by Survival Status")
plt.xlabel("Survived")
plt.ylabel("Age")
```

```python
# scatter plot of age versus fare paid to explore the relationship between these two variables
sns.scatterplot(x=titanic_data["Age"], y=titanic_data["Fare"])
plt.title("Fare Paid and Age of Passengers")
plt.xlabel("Age")
```

```python
# bar chart comparing the number of male and female passengers by passenger class
```

```python
# heatmap showing the total number of survivors for each class broken down by age
titanic_data["Age Group"] = pd.cut(titanic_data["Age"], bins=[0, 18, 80])
titanic_data.head()
sns.heatmap(titanic_data.pivot_table(index="Age Group", columns="Pclass", values="Survived", aggfunc=sum), annot=True, fmt=".0f")
plt.title("Number of Survivors by Age Group and Passenger Class")
```

```python
# oldest survivor on titanic
titanic_data.sort_values(by="Age", ascending=False).head(1)
```

```python
# stacked bar chart comparing the number of survivors and non-survivors by passenger class
sns.countplot(x=titanic_data["Pclass"], hue=titanic_data["Survived"])
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Number of Passengers")
```
